## speed and weight of different buses given above

```python
plt.title ("Speed vs weight")
plt.xlabel ("Speed (km/s)")
plt.ylabel ("weight (kg)")
plt.scatter(speed,weight)
```

```python
# check linear regression using the stats module

stats.linregress(speed,weight)
```

```python

## destructuring
slope,intercept, r, p, stderr = stats.linregress(speed,weight)
```

```python
slope
```

```python
# y = mx + c
## createna function to get new values of y

def myFunc (x):
    return slope * x + intercept

# run each value of x through the function
# this will give us new values of y

new_y = list(map(myFunc, speed))

# draw scatter plot
plt.scatter (speed,weight)

# draw the line
plt.xlabel('Bus Speed')
plt.ylabel('weight')
plt.plot (speed, new_y)
```

```python
# find the weight of a bus moving at 93

w = myFunc (93)
print (w)
```

```python
# the r value (-0.43...) show us that linear regression does not give the best fit
```

```python
## ploynomial regression
 
model = np.poly1d (np.polyfit (speed,weight,3))
```

```python
# formular
#
# z = x-u/ S.D
#
# z = scaled value
# x = feature
# u = mean
# S.D = standard deviation
```

```python
# create a function to resolve 
def zscore_norm(x,u,std):
    z = (x-u)/std
    return z
```

```python
# find the mean and S.D for each column

Score1_mean = df.Score1.mean()
Score1_std = df.Score1.std()

print("mean: ", Score1_mean,'\n',"S.D:",
     Score1_std)
```

```python
# Score2

Score2_mean = df.Score2.mean()
Score2_std = df.Score2.std()

print("mean: ", Score2_mean,'\n',"S.D",
     Score2_std)
```

```python
def greet():
    now = datetime.datetime.now()
    currentHour = now.hour
    msg = "hello "
    msg2 = ""
    
    # check the time of day
    if currentHour < 12:
        msg2 = "Good morning"
    elif currentHour < 18:
        msg2 = "Good afternoon"
    else:
        msg2 = "Good evening"
    
    result = msg + msg2
    print(result)
greet()        
```
